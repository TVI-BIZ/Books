You've just been hired at a real estate investment firm and they would like you to build a model for pricing houses. You are given a dataset that contains data for house prices and a few features like number of bedrooms, size in square feet and age of the house. Let's see if you can build a model that is able to predict the price. In this exercise we extend what we have learned about linear regression to a dataset with more than one feature. Here are the steps to complete it:

1. load the dataset ../data/housing-data.csv
- plot the histograms for each feature
- create 2 variables called `X` and `y`: `X` shall be a matrix with 3 columns (sqft,bdrms,age) and `y` shall be a vector with one column (price)
- create a linear regression model in Keras with the appropriate number of inputs and output
- split the data into train and test with a 20% test size
- train the model on the training set and check its accuracy on training and test set
- how's your model doing? Is the loss growing smaller?
- try to improve your model with these experiments:

```python
#load the dataset ../data/housing-data.csv
df_real= pd.read_csv('../data/housing-data.csv')
df_real.head(3)
```

```python
#plot the histograms for each feature
df_real.hist(bins=4)
```

```python
#create 2 variables called X and y: 
#X shall be a matrix with 3 columns (sqft,bdrms,age) 
#and y shall be a vector with one column (price)
X_real = df_real[['sqft','bdrms','age']]
y_real = df_real.price
y_real.head()
#X.head()
# print('   ',len(X_real))
# print('   ',len(y_real))
```

```python
#create a linear regression model in Keras with the appropriate number of inputs and output
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam, SGD, RMSprop
model_real = Sequential()
model_real.add(Dense(1, input_shape=(3,)))
model_real.summary()
```

```python
X_trainR, X_testR, y_trainR, y_testR = train_test_split(X_real,y_real,test_size=0.2)
```

```python
#rain the model on the training set and check its accuracy on training and test set
from sklearn.metrics import r2_score

model_real.compile(Adam(lr=0.8), 'mean_squared_error')
model_real.fit(X_trainR, y_trainR, epochs=25)

y_train_pred_real = model_real.predict(X_trainR)
y_test_pred_real = model_real.predict(X_testR)

r_ = r2_score(y_trainR, y_train_pred_real)
print("R2 score on Train set is:\t{:0.3f}".format(r_))

r_ = r2_score(y_testR, y_test_pred_real)
print("R2 score on Test set is:\t{:0.3f}".format(r_))
```

```python
def pretty_cm(y_true, y_pred, labels=["False", "True"]):
    cm = confusion_matrix(y_true, y_pred)
    pred_labels = ['Predicted '+ l for l in labels]

    df = pd.DataFrame(cm,
                      index=labels,
                      columns=pred_labels)
    return df
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
y_hr_pred = model_HR.predict(X)
y_hr_class_pred = y_hr_pred > 0.5

precision = precision_score(y, y_hr_class_pred)
print("Precision:\t{:0.3f}".format(precision))

recall = recall_score(y, y_hr_class_pred)
print("Recall:  \t{:0.3f}".format(recall))

f1 = f1_score(y, y_hr_class_pred)
print("F1 Score:\t{:0.3f}".format(f1))
print('   ',len(X))
print('   ',len(y_hr_class_pred))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
def build_logistic_regr():
    model = Sequential()
    model.add(Dense(1, input_dim=18, activation='sigmoid'))
    model.compile(Adam(lr=0.5),
                  'binary_crossentropy',
                  metrics=['accuracy'])
    return model

model = KerasClassifier(build_fn=build_logistic_regr,
                        epochs=10, verbose=0)

cv = KFold(3, shuffle=True)
scores = cross_val_score(model, X, y, cv=cv)
```

```python
m = scores.mean()
s = scores.std()
print("Cross Validation accuracy:",
      "{:0.4f} ± {:0.4f}".format(m, s))
```
