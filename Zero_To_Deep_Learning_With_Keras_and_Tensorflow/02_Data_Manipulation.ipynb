### Exercise 3

- plot the histogram of the heights for males and for females on the same plot
- use `alpha` to control transparency in the plot comand
- plot a vertical line at the mean of each population using `plt.axvline()`
- bonus: plot the cumulative distributions

```python
df_weight.head()
```

```python
#df[df['Age'] > 79]
import matplotlib.pyplot as plt
import numpy as np
title = 'Male and Female'
df_male = df_weight[df_weight['Gender'] == 'Male']
df_female = df_weight[df_weight['Gender'] == 'Female']
df_female['Height'].hist(alpha=0.5,label='female')
df_male['Height'].hist(alpha=0.8,label='male')
plt.axvline(df_male['Height'].mean(), color='red')
plt.axvline(df_female['Height'].mean(), color='black')
plt.legend()
plt.show()
```
